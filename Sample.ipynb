```python
import matplotlib.pyplot as plt
from skimage import io
```

```python
PATH_TO_AUTHENTIC_IMG = os.getcwd() + '/ImSpliceDataset/Au-S/AU_S_023.bmp'
PATH_TO_SPLICED_IMG = os.getcwd() + '/ImSpliceDataset/Sp-S/SP_S_023.bmp';
PATH_TO_LOCALISED_IMAGE_EUC = os.getcwd() + '/Output/Sp-S/SP_S_023_euclidean.bmp'
PATH_TO_LOCALISED_IMAGE_HELL = os.getcwd() + '/Output/Sp-S/SP_S_023_hellinger.bmp'
im1 = io.imread(PATH_TO_AUTHENTIC_IMG)
im2 = io.imread(PATH_TO_SPLICED_IMG)
im3 = io.imread(PATH_TO_LOCALISED_IMAGE_EUC)
im4 = io.imread(PATH_TO_LOCALISED_IMAGE_HELL)
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax1.title.set_text('Actual Image')
io.imshow(im1)
ax2 = fig.add_subplot(2,2,2)
ax2.title.set_text('Spliced Image')
io.imshow(im2)
ax3 = fig.add_subplot(2,2,3)
ax3.title.set_text('Euclidean Distance')
io.imshow(im3)
ax4 = fig.add_subplot(2,2,4)
ax4.title.set_text('Hellinger Distance')
io.imshow(im4)
```
