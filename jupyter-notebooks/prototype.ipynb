## Simple Linear Iterative Clustering (SLIC) for Superpixel Segmentation

For information re SLIC algorithm, follow [ Radhakrishna Achanta, Appu Shaji, Kevin Smith, Aurelien Lucchi, Pascal Fua, and Sabine SÂ¨usstrunk, SLIC Superpixels, EPFL Technical
Report 149300, June 2010](https://www.iro.umontreal.ca/~mignotte/IFT6150/Articles/SLIC_Superpixels.pdf)

````
# import the necessary packages

from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io
import matplotlib.pyplot as plt
````

````
# load images and store in dictionary as filename: image

path = './images/'
filenames = sorted([
    'raptors.png',
    'lena.png'
])

image_dict = {}
for filename in filenames:
    image_dict[filename] = img_as_float(io.imread(path + filename))
````

````
# show images

for filename, image in image_dict.items():
    plt.imshow(image)
    plt.title(filename)
    plt.show()
````

````
segments_dict = {}

segments_n_list = [100, 200, 300]

# loop over list of images running SLIC (k-means clustering) on each
for filename, image in image_dict.items():
    # create figure for imshow
    fig = plt.figure()
    fig.tight_layout()
    fig.set_dpi(350)

    segments_dict[filename] = {}

    # loop for values of segments_n
    for i, segments_n in enumerate(segments_n_list):
        #apply SLIC and (approximately) extract segments_n segments
        segments = slic(image, n_segments=segments_n, sigma=5)
        segments_dict[filename][segments_n] = segments

        # add segmented images as subplots
        ax = fig.add_subplot(1, len(segments_n_list), i + 1, title=segments_n)
        ax.imshow(mark_boundaries(image, segments))
        io.imsave(f'outs/{filename}-{segments_n}.png', mark_boundaries(image, segments))
        ax.axis(False)
    plt.subplots_adjust(left=0, right=0.5)
    plt.show()
````

### Assembled data format

```typescript
image_data_dict = {
    'filename' = {
        'image': Array[Array[]],                # image 2D array
        'segments': {
            [key: Int]: Array[Array[Array[]]]   # segments count: array of 2D segments arrays
        }
    }
}
```

````
# assemble data into a dictionary in format

image_data_dict = {}
for filename in filenames:
    image_data_dict[filename] = {
        'image': image_dict[filename],
        'segments': segments_dict[filename]
    }
````
